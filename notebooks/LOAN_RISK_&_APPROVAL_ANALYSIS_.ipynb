```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('loan_risk_prediction_dataset_raw.csv')
```

```python
# Print unique values to check for inconsistent spellings or trailing spaces
print("Unique Education Categories:", df['Education'].unique())
print("Unique Cities:", df['City'].unique())
print("Unique Employment Types:", df['EmploymentType'].unique())
```

```python
# Convert negative values to NaN so they can be imputed properly
df.loc[df['Income'] < 0, 'Income'] = np.nan
df.loc[df['LoanAmount'] < 0, 'LoanAmount'] = np.nan
df.loc[df['CreditScore'] < 0, 'CreditScore'] = np.nan
```

```python
# Impute numerical columns with their medians
df['Income'] = df['Income'].fillna(df['Income'].median())
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
df['CreditScore'] = df['CreditScore'].fillna(df['CreditScore'].median())

# Impute and clean categorical columns
df['Education'] = df['Education'].fillna('Unknown').str.title()
df['EmploymentType'] = df['EmploymentType'].fillna('Unknown').str.title()
```

```python
print("--- FINAL DATA QUALITY CHECK ---")
print(df.isnull().sum())
```

```python
from google.colab import files

# Save to CSV
df.to_csv('loan_risk_data_cleaned.csv', index=False)

# Download to your local machine
files.download('loan_risk_data_cleaned.csv')
```
